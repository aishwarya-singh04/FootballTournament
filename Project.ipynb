1. Read the data set

```python
dsLaliga = pd.read_csv('Laliga.csv',header=1)
```

```python
dsLaliga.head(5)
```

Replace dashes with 0

```python
dsLaliga.replace('-',0, inplace=True)
```

```python
#"-" replaced to 0.
dsLaliga.tail(5)
```

2. Print all the teams started playing between 1930-1980

```python
debutYear = dsLaliga[dsLaliga['Debut'].astype(str).str[:4].astype(int).between(1930, 1980)]
debutYear[['Team','Debut']]
```

3. Print the list of teams which came Top 5 in terms of points

```python
dsLaliga[['Team','Points']].sort_values(by= 'Points', ascending= False).head(5)
```

4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.

```python
def GoalDiffCount():   
     dsLaliga['GoalDiffCount'] = dsLaliga['GoalsFor'].astype(int)-dsLaliga['GoalsAgainst'].astype(int)
     return dsLaliga[['Team','GoalDiffCount']].sort_values(by = 'GoalDiffCount', ascending=False)
GoalDiffCount()
```

```python
#Team with maximum goal count
GoalDiffCount().head(1)
```

```python
#Team with minimum goal count
GoalDiffCount().tail(1)
```

 5. Create a new column with name “Winning Percent” and append it to the data set
Percentage of Winning = (GamesWon / GamesPlayed)*100

```python
dsLaliga['Winning Percent'] = (dsLaliga['GamesWon'].astype(int)/dsLaliga['GamesPlayed'].astype(int)) *100
dsLaliga['Winning Percent']
```

```python
#If there are any numerical error, replace it with 0%
dsLaliga['Winning Percent'].fillna(0, inplace=True)
dsLaliga['Winning Percent']
```

```python
#Print the top 5 teams which has the highest Winning percentage
dsLaliga[['Team', 'Winning Percent']].head(5)
```

6. Group teams based on their “Best position” and print the sum of their points for all positions

```python
grouped_BestPosition = dsLaliga[['Team','Points','BestPosition']].groupby('BestPosition')
grouped_BestPosition
```

```python
grouped_BestPosition.sum()
```
